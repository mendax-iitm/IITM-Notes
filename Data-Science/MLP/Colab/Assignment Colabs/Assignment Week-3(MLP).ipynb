#Practice Assignments

```python
import numpy as np
```

```python
from sklearn.datasets import fetch_california_housing
```

```python
features, labels = fetch_california_housing(as_frame=True, return_X_y=True)
```

```python
features.shape
```

```python
features.isna().sum().sum()
```

```python
from sklearn.model_selection import train_test_split
train_features, test_features, train_labels, test_labels = train_test_split(features, labels, random_state=0,shuffle=False,test_size=0.2)
```

```python
test_features
```

```python
train_features
```

```python
np.mean(train_labels)
```

```python
np.median(test_labels)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
lin_reg_pipeline = Pipeline([("feature scaling",StandardScaler()),("linear regression",LinearRegression())])
lin_reg_pipeline.fit(train_features,train_labels)
```

```python
lin_reg_pipeline[-1].coef_
```

```python
lin_reg_pipeline[-1].intercept_
```

# Graded

```python
lin_reg_pipeline.score(test_features,test_labels)
```

```python
from sklearn.metrics import mean_squared_error
test_predictions = lin_reg_pipeline.predict(test_features)
mean_squared_error(test_labels,test_predictions, squared=False)
```

```python
from sklearn.metrics import explained_variance_score
test_predictions = lin_reg_pipeline.predict(test_features)
explained_variance_score(test_labels,test_predictions)
```

```python
from sklearn.metrics import max_error
max_error(test_labels,test_predictions)
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(test_labels,test_predictions)
```

```python
from sklearn.metrics import mean_squared_error
test_predictions = lin_reg_pipeline.predict(test_features)
mean_squared_error(test_labels,test_predictions)
```

```python
from sklearn.linear_model import SGDRegressor
sgd_reg_pipeline = Pipeline([("feature scaling",StandardScaler()),("SGD",SGDRegressor(random_state=0))])
sgd_reg_pipeline.fit(train_features,train_labels)
```

```python
sgd_reg_pipeline[-1].intercept_
```
